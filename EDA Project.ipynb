```python
df3.query('race_length == "50mi"') \
   .groupby("athlete_age")['athlete_average_speed'] \
   .agg(['mean', 'count']) \
   .sort_values('mean', ascending=True) \
   .query('count > 9') \
   .head(20)
```

```python
#Season for the data -> Slower in summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12 -2
#split between two decimals
```

```python
df3['race_month'] = df3 ['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3 [ 'race season' ] = df3 ['race_month'].apply(lambda x:"winter" if x > 11 else "fall" if x>8 else 'summer' if x > 5 else 'spring' if x>2 else 'winter')
```

```python
df3.tail(25)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count'])
```
